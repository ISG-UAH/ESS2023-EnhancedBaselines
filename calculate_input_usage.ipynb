```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
src_file = 'results/expressions_elasticnetAR_1d_24d.csv'
```

```python
daily_inputs = len(['Br', 'Bt', 'Bn', 'V', 'N', 'T', 'Br', 'Bt', 'Bn', 'V', 'N', 'T'])*24
daily_inputs
```

```python
expressions_pd = pd.read_csv(src_file, index_col = 0)
expressions_pd
```

```python
expr = expressions_pd.latex.values
expr
```

```python
inp_days = np.vectorize(lambda x: eval(x.replace('1h_', '').replace('(', '').replace(')', '').split(',')[0]))(expressions_pd.index.values)
inp_days
```

```python
n_inputs = inp_days * daily_inputs
n_inputs
```

```python
rhs = np.vectorize(lambda x: x.split('=')[1])(expr)
rhs
```

```python
for e in rhs:
    print(e.split('+'))
    print()
```

```python
used = np.vectorize(lambda x: len(x.split('+'))-1)(rhs)
used
```

```python
prop = used/n_inputs
percentage = prop*100
prop, percentage
```

```python
print(f'Avg usage: {np.mean(percentage):.2f}%')
print(f'Std dev of usage: {np.std(percentage):.2f}%')
```

```python
indices = np.vectorize(lambda x: x.replace('1h_', ''))(expressions_pd.index.values)
indices
```

```python
processed_idx = []
for i in indices:
    processed_idx.append(eval(i))
processed_idx
```

```python
usage_percentage_pd = pd.DataFrame(percentage, columns=['Usage %'], index=processed_idx)
usage_percentage_pd
```

```python
sns.histplot(usage_percentage_pd, x='Usage %', kde=True, stat='probability');
```

```python
inp_days = np.array(processed_idx)[:, 0]
inp_days
```

```python
lead_times =  np.array(processed_idx)[:, 1]
lead_times
```

```python
usage_percentage_pd['Input days'] = inp_days
usage_percentage_pd['Lead time (days)'] = lead_times
usage_percentage_pd
```

```python
for in_days in np.unique(usage_percentage_pd['Input days']):
    print(f'Input size = {in_days} days')
    print(f'Avg usage: {np.mean(usage_percentage_pd[usage_percentage_pd["Input days"]==in_days].values):.2f}%')
    print(f'Std dev of usage: {np.std(usage_percentage_pd[usage_percentage_pd["Input days"]==in_days].values):.2f}%')
    print()
```

```python
ld_time_mean = []
ld_time_std = []
for lead in np.unique(usage_percentage_pd['Lead time (days)']):
    print(f'Lead time = {lead} days')
    avg = np.mean(usage_percentage_pd[usage_percentage_pd["Lead time (days)"]==lead].values)
    print(f'Avg usage: {avg:.2f}%')
    std = np.std(usage_percentage_pd[usage_percentage_pd["Lead time (days)"]==lead].values)
    print(f'Std dev of usage: {std:.2f}%')
    print()
    
    ld_time_mean.append(avg)
    ld_time_std.append(std)
```

```python
plt.plot([1, 2, 3, 4], ld_time_mean)
plt.title('Proportion of input used on average per lead time')
plt.xlabel('Lead time (days)')
plt.ylabel('% of input used');
```

```python
plt.plot([1, 2, 3, 4], ld_time_std)
plt.title('$\sigma$ of proportion of input used per lead time')
plt.xlabel('Lead time (days)')
plt.ylabel('% of input used');
```
