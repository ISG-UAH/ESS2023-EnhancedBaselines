```python
import pandas as pd
import numpy as np
import scipy as scp
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm.notebook import tqdm
from sklearn.preprocessing import MinMaxScaler 
from timeit import timeit

import os
```

```python
DEBUG = False
```

```python
if not os.path.isdir('figs'):
    os.mkdir('figs')
```

```python
models = {
    #'Elasticnet AR': [(5, 1), (5, 2), (5, 3), (5, 4)],
    #'Linear SVR': [(5, 1), (5, 2), (5, 3), (5, 4)],
    'SVR': [(3, 1), (4, 2), (4, 3), (5, 4)]
}
```

```python
t_horizons = ['1d', '2d', '3d', '4d', '27d']
```

```python
def plot_and_save(real, pred, model_name, configuration):
    r = real.loc['2017-01-01':'2018-01-01'].interpolate(method='time')
    r.plot(figsize=(15, 5), title='')
    p = pred.reindex(r.index)['V']
    if DEBUG:
        display(r)
        display(p)
    p.plot(style='--', title='')
    plt.title('')
    plt.suptitle('')
    plt.xlabel('')
    plt.ylabel('Bulk speed (km$\cdot$s$^{-1}$)')
    plt.legend(['Observation', model_name])
    if 'persistence' in model_name:
        if not ('27') in model_name:
            plt.title(f'Lead time: {configuration[0]} day{"" if "1" in model_name else "s"}')
        t_file = f"figs/{model_name}.png"
    else:
        t_file = f"figs/{model_name.replace(' ', '_')}_{configuration[0]}d_{configuration[1]}d.png"
    plt.savefig(t_file)
    plt.show()    
    plt.close()
```

```python
wind_data = pd.read_csv('data/omni_1hr_clean.csv', parse_dates=[0], index_col=0)
wind_data
```

```python
for lead_time in t_horizons:
    raw_time = eval(lead_time.replace('d', ''))
    raw_hours = raw_time*24
    shifted = wind_data.shift(raw_hours).loc['2017-01-01':'2018-01-01'].interpolate(method='time')
    plot_and_save(wind_data['V'], shifted, f'{raw_time} day{"s" if raw_time > 1 else ""} persistence', (raw_time, 'na'))
```

```python
for m_name in models.keys():
    n = m_name.split(' ')
    if len(n) == 2:
        n[0] = n[0].lower()
        n[1] = n[1].upper()
        n = n[0]+n[1]
    else:
        n = n[0]
    for m_config in models[m_name]:
        t_file = f'results/{n}_predictions_{m_config[0]}_{m_config[1]}.txt'
        try:
            preds = pd.read_csv(t_file, parse_dates=[1], index_col=0).set_index('Timestamp')
            preds['V'] = preds['Prediction'].apply(lambda x: eval(x)[0])
            if DEBUG:
                display(preds)
            plot_and_save(wind_data['V'], preds, m_name, m_config)
        except Exception as e:
            if DEBUG:
                print(e)
            plt.close()
            continue
```
